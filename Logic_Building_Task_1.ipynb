```python
marks = [45, 78, 90, 33, 60]
p=0
f=0
for mark in marks:
  if mark>=50:
    p+=1
  else:
    f+=1
print("total Passed Students",p)
print("total Failed Students",f)
```

```python
names = [" Alice ", "bob", " CHARLIE "]
cleaned = []
for name in names:
    c= name.strip().lower()#remove spaces and convert to lower case
    cleaned.append(c)
print("Cleaned Names List:", cleaned)
```

```python
message = ["Hi", "Welcome to the platform", "OK"]
for m in message:
  length=len(m)#length of message
  print("Message:",m)
  print("Length:",length)
  if length>10:#check if message is longer than 10 characters
    print("Status:Long Message")
  else:
    print("Status:Short Message")
```

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
c=0
for l in logs:
  if l=="ERROR":
    c+=1
print("Error Count:",c)
```
